```python
dfinfo = pd.read_csv('C:/Users/fid24/downloads/superhero_info - superhero_info.csv')
dfinfo.head()
```

```python
dfinfo[['Hero','Publisher']] = dfinfo['Hero|Publisher'].str.split('|',expand=True)
dfinfo.head(2)
```

```python
dfinfo = dfinfo.drop(columns=['Hero|Publisher'])
dfinfo.head(2)
```

```python
meas = dfinfo.loc[0,"Measurements"]
print(type(meas))
meas
```

```python
meas = meas.replace("'",'"')
meas
```

```python
fixed_meas = json.loads(meas)
print(type(fixed_meas))
fixed_meas
```

```python
## use .str.replace to replace all single quotes
dfinfo['Measurements'] = dfinfo['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
dfinfo['Measurements'] = dfinfo['Measurements'].apply(json.loads)
dfinfo['Measurements'].head()
```

```python
height_weight = dfinfo['Measurements'].apply(pd.Series)
height_weight
```

```python
dfinfo = pd.concat((dfinfo, height_weight), axis = 1)
dfinfo.head(2)
```

```python
dfinfo['Height'].str.split(' ',expand=True)
```

```python
dfinfo[['Height_in_CM','cm']] = dfinfo['Height'].str.split(' ',expand=True)
dfinfo.head(2)
```

```python
dfinfo['Weight'].str.split(' ',expand=True)
```

```python
dfinfo[['Weight_in_KG','kg']] = dfinfo['Weight'].str.split(' ',expand=True)
dfinfo.head(2)
```

```python
dfinfo = dfinfo.drop(columns=['Measurements','cm','kg','Height','Weight'])
dfinfo.head(2)
```

```python
dfinfo['Height_in_CM'] = dfinfo['Height_in_CM'].astype(float)
```

```python
dfinfo['Weight_in_KG'] = dfinfo['Weight_in_KG'].astype(float)
```
